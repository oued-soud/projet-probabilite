### telechargement du livre 

```python
import string
import urllib.request
import nltk
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np



x = urllib.request.urlopen('https://r-stat-sc-donnees.github.io/LesMiserables1.txt')
livre=str(x.read())
```

## suppression de la ponctuation, les majuscules / minuscules et divisez le texte en jetons individuels (mots)

```python

livre = livre.lower()
nltk.word_tokenize(livre)
tokenizer = nltk.RegexpTokenizer(r'\w+')
livre = tokenizer.tokenize(livre)
#print(livre)
```

## Pour les  calcule le nombres respectifs des mots (un, et, le, il, est) au fur et Ã  mesure que le livre progresse

```python
def nombre_mots(mot):
    mots=[]
    a=0
    for i in range(len(livre)):
        if(livre[i]==mot):
            a+=1
            mots.append(a)
        else:
            mots.append(0)
    return a,mots



n_un, un = nombre_mots('un')
n_et, et = nombre_mots('et')
n_le, le = nombre_mots('le')
n_il, il = nombre_mots('il')
n_est, est = nombre_mots('est')



    

tab = [n_un,n_et, n_le,n_il,n_est]

    
nom = ["un","et", "le","il","est"]      

mots = {'nom': nom,
         'nombre': tab,
         
         }

data = pd.DataFrame(mots)
print(data)  
```

### Tracons les proportions nmot ris {i sur le document dans un seul graphique.

```python
mots = [un,et, le,il,est]
for i in range (len(mots)):
    plt.plot(np.arange(len(mots[i])), mots[i])
    plt.title(nom[i] )
    plt.show()
    
```
